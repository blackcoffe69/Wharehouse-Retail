# Analyzing the value count of each section using bar chart

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
pip install plotly
```

```python
sales_by_month = Ware_house.groupby("MONTH")["WAREHOUSE SALES"].sum().reset_index()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=sales_by_month, x="MONTH", y="WAREHOUSE SALES", marker='o', color='blue')
```

```python
print(sales_by_month)
# month 7 shows the highest Wharehouse sales of 1109919.19 sales
# the plot shows month 12 with the lowest sales 306957.22 sales
```

```python
sales_pivot = Ware_house.pivot_table(index='YEAR', columns='ITEM TYPE', aggfunc='size', fill_value=0)
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=sales_pivot.stack().reset_index(), x='YEAR', y=0, hue='ITEM TYPE', palette='muted', ci=None)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales by Year and Item Type')
plt.legend(title='Item Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
print(sales_pivot)
```

# Aggregating Supplier value count

```python
supplier_counts = Ware_house["SUPPLIER"].value_counts().reset_index()
supplier_counts.columns = ["SUPPLIER", "Count"]
top_50_suppliers = supplier_counts.head(50)
# getting the first (50) supplies to show only because the number of suppliers exceed what the bar chart can contain
```

```python
plt.figure(figsize = (9, 8))
sns.barplot(data = top_50_suppliers, x = "Count", y = "SUPPLIER")
plt.title("SUPPLIER BY COUNT")
plt.xlabel("Count")
plt.ylabel("SUPPLIER")
plt.tight_layout()
plt.show()
```

```python
print(top_50_suppliers)
```

# Aggregating the Warehouse sales per year

```python
WS = Ware_house.sort_values("WAREHOUSE SALES").reset_index(drop = True)
```

```python
WS
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=WS, x = "YEAR", y = "ITEM TYPE", marker = "o")
plt.title("ITEM TYPE by Year")
plt.xlabel("YEAR")
plt.ylabel("ITEM TYPE")
plt.xticks(Ware_house["YEAR"])
plt.tight_layout()
plt.show()
```

```python
print(WS)
# It shows that their was reduction in sales in 2018
# The increase in sales starts to show after then with the peek period till 2020
```

```python
# Displaying sales by year and item type for retail sales
sales_by_item_year = Ware_house.groupby(["ITEM TYPE", "YEAR"])["RETAIL SALES"].sum().unstack().fillna(0)
```

```python
# Plotting
plt.figure(figsize=(12, 8))
sales_by_item_year.plot(kind="bar", stacked=True)
plt.legend(title="YEAR", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# this shows the stack bar chart for the Retail sales by year with 2019 having the highest sales of 262.4797k for beer,357.0774k for liquor,13.98473k for Non alcholic,325.1128k for Wine
# the year 2018 has the lowest sales with beer having Retails sales of 35.24664k,Liqour having 57.43218k,Wine 59.20211k
```
